```python
import re
```

```python
def fmt_field(s):
    s = re.sub(r'[- "\n\r]*', '', s)
    if s[0].isdigit():
        s = '_' + s
    return s

def from_csv(filename, delimiter=',', quotechar='"'):
    with open(filename) as csvfile:
        reader = csv.reader(csvfile, delimiter=delimiter, quotechar=quotechar)

        # ignore header
        next(reader)

        return tuple(((fmt_field(field), fmt) for field, _, _, fmt, *_ in reader))

from_csv('contrib/LOOP2.csv')
```
